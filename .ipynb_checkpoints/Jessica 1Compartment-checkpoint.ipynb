```python
import numpy as np
import scipy.io as spio
from scipy.integrate import solve_ivp
##%matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib import rc, rcParams

Cm = 1.0
gCl = 0.3
tau = 10.0/3.0
V0 = -68.0
A = 4*(np.pi)*(10**-3)**2
Vcl = -68.0
t1 = 2.5
t2 = 22.5

def Istim(t):
    if (t < t1 or t > t2):
        return 0
    else:
        return 10**-5

# x axis values
t_span = (0., 40.)
# corresponding y axis values
def model1(t, y):
    V = y
    dVdt = Istim(t)/(A*Cm) + (Vcl-V)/tau
    return [dVdt]

sol = solve_ivp(model1, t_span, [V0])
# plotting the points 
plt.plot(sol.t, sol.y[0, :], 'b', label='Vs(t)')
  
# naming the x axis
plt.xlabel('time(ms)')
# naming the y axis
plt.ylabel('membrane potential (mV)')

  
# function to show the plot
plt.show()

```

```python
import math
def fun (t, t1, t2): 
    if t<t1:
        return Vcl
    elif t>=t1 and t<=t2:
        return Vcl + 10**-5*(t2-t1)*tau*(1-math.e**((t1-t)/tau))/((t2-t1)*A*Cm)
    else:
        return Vcl + 10**-5*(t2-t1)*tau*((math.e**((t2-t)/tau))-math.e**((t1-t)/tau))/((t2-t1)*A*Cm)

vfun = np.vectorize(fun)

t = np.linspace(0, 40, 5000)    
y = vfun(t, t1, t2)

plt.plot(t, y, '-')
plt.show()
```
